```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'])
```

```python
df.replace({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4},inplace=True)
```

```python
df
```

```python
df['Total_Stops'].value_counts()
```

```python
plt.title('No. of flights stopage')
plt.hist(df['Total_Stops'])
plt.xlabel('Stopages')
plt.ylabel('no. of flights')
plt.xticks(df['Total_Stops'].unique())
plt.show()
```

```python
plt.title('No. of flights stopage')
plt.bar(df['Airline'],df['Total_Stops'])
plt.xlabel('Airline')
plt.ylabel('Total stopages')
plt.xticks(rotation=90)
plt.show()
```

```python
df
```

```python
g=df.groupby(['Airline'])['Price'].agg(['sum','max']).reset_index().sort_values(by='sum',ascending = False)
g
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.bar(g['Airline'],g['sum'],label='overall profit',color='r')
plt.xticks(rotation=90)
plt.legend()

plt.subplot(1,2,2)
plt.bar(g['Airline'],g['max'],label='expensive and cheap',color='k')
plt.xticks(rotation=90)
plt.legend()
```
